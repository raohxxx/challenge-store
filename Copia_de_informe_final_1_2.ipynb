# 4. Productos m√°s y menos vendidos

```python
producto1=tienda1['Producto'].value_counts()
producto2=tienda2['Producto'].value_counts()
producto3=tienda3['Producto'].value_counts()
producto4=tienda4['Producto'].value_counts()

consolidado2 = pd.concat([producto1, producto2, producto3, producto4], axis=1)
consolidado2.columns = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
#print(consolidado2)
#consolidado2['Tienda 1'].idxmax()
for x in consolidado2.columns:
    valor_maximo=consolidado2[x].max()
    filas_maximas = consolidado2[consolidado2[x] == valor_maximo]
    indices_maximos = filas_maximas.index.tolist()
    print(f' el producto mas vendido de la {x} es o son:  {indices_maximos}')
    valor_minimo=consolidado2[x].min()
    filas_minimas = consolidado2[consolidado2[x] == valor_minimo]
    indices_minimos = filas_minimas.index.tolist()
    print(f' el producto menos vendido de la {x} es o son:  {indices_minimos}\n')


consolidado2
```

```python
# Suponiendo que tienes los DataFrames tienda1, tienda2, tienda3, tienda4
# con la columna 'Producto' que contiene el nombre de cada producto

# Combinar los DataFrames de las tiendas
all_stores = pd.concat([tienda1, tienda2, tienda3, tienda4], keys=['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'])
# Reiniciar el √≠ndice para evitar el error 'level_0'
all_stores = all_stores.reset_index()
# Agrupar por tienda y producto y contar las ventas
product_counts = all_stores.groupby(['level_0', 'Producto'])['Producto'].count().reset_index(name='Ventas')

# Crear la figura y los subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Iterar sobre las tiendas y los subplots
for i, store in enumerate(['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']):
    row = i // 2  # Fila del subplot
    col = i % 2   # Columna del subplot
    ax = axs[row, col]

    # Filtrar los datos de la tienda actual
    store_data = product_counts[product_counts['level_0'] == store]

    # Crear el gr√°fico de barras
    ax.bar(store_data['Producto'], store_data['Ventas'])

    # Configurar el t√≠tulo y las etiquetas
    ax.set_title(store)
    ax.set_xlabel('Producto')
    ax.set_ylabel('Ventas')

    # Rotar las etiquetas del eje x para mejor legibilidad
    plt.setp(ax.get_xticklabels(), rotation=85, ha='right')

# Ajustar el espaciado entre los subplots
plt.tight_layout()

# Mostrar el gr√°fico
plt.show()
```

##Informaci√≥n general:

**Ubicaci√≥n de las ventas geograficamente**:
 Los gr√°ficos muestran la ubicaci√≥n geogr√°fica de las ventas en un mapa. La posici√≥n de cada punto representa la latitud y longitud de una venta. Los puntos de diferentes colores representan las ventas por tiendas 1, 2, 3 y 4, respectivamente, permitiendo ver la distribuci√≥n espacial de cada venta por tienda.
Concentraci√≥n espacial:
 Se puede observar si las ventas est√°n concentradas en √°reas geogr√°ficas espec√≠ficas o si est√°n distribuidas de forma m√°s dispersa. La concentraci√≥n de puntos en una zona podr√≠a indicar una mayor presencia de la marca en esa regi√≥n.
Superposici√≥n:
 Si hay superposici√≥n de puntos de diferentes colores en la misma √°rea geogr√°fica, significa que las diferentes tiendas comparten clientes en esa zona. Esto podr√≠a ser una se√±al de competencia o una estrategia para cubrir un mercado en particular.
Zonas sin presencia:
 Las √°reas sin puntos indican zonas donde las tiendas no tienen presencia f√≠sica. Esto podr√≠a sugerir oportunidades para expandirse o indicar que la empresa no est√° interesada en llegar a esas zonas por alguna raz√≥n.
Informaci√≥n sobre el comportamiento de las ventas o clientes:

Relaci√≥n con otras variables:
 Los gr√°ficos de dispersi√≥n geogr√°fica pueden combinarse con otras variables (como el volumen de ventas, el tipo de cliente, etc.) para obtener informaci√≥n adicional. Por ejemplo, podr√≠as agregar un mapa de calor para mostrar las zonas con mayor densidad de ventas.
Patrones geogr√°ficos:
 Si observas patrones geogr√°ficos particulares en la distribuci√≥n de los puntos, podr√≠as usar esa informaci√≥n para segmentar el mercado o tomar decisiones de marketing m√°s efectivas. Por ejemplo, se podr√≠a dirigir tus campa√±as publicitarias a regiones con una mayor concentraci√≥n de ventas.
Zonas de influencia:
 Los gr√°ficos pueden ayudar a identificar la zona de influencia de cada tienda. Esto es √∫til para determinar el alcance de la tienda en el mercado, identificar la competencia en la zona y optimizar la log√≠stica de entrega.
Ejemplo:

Si observas una alta concentraci√≥n de puntos de la tienda 1 y la tienda 2 en un √°rea espec√≠fica, podr√≠as concluir que:

Las tiendas 1 y 2 tienen una mayor presencia en esa zona, lo cual podr√≠a indicar un mercado potencialmente m√°s atractivo.
Las tiendas 1 y 2 podr√≠an estar compitiendo por los mismos clientes en esa zona.
Limitaciones:

Los gr√°ficos de dispersi√≥n geogr√°fica solo ofrecen una vista est√°tica de la distribuci√≥n espacial. Para un an√°lisis m√°s profundo, se podr√≠an usar otras herramientas, como mapas interactivos, an√°lisis de redes, etc.
La informaci√≥n derivada de los gr√°ficos debe ser interpretada en conjunto con otros datos para obtener una visi√≥n m√°s completa del mercado y las tiendas.
Es importante tener en cuenta que la interpretaci√≥n de la informaci√≥n de los gr√°ficos de dispersi√≥n geogr√°fica depender√° del contexto espec√≠fico de tu negocio y los objetivos de tu an√°lisis.
