# 4. Productos más y menos vendidos

```python
producto1=tienda1['Producto'].value_counts()
producto2=tienda2['Producto'].value_counts()
producto3=tienda3['Producto'].value_counts()
producto4=tienda4['Producto'].value_counts()

consolidado2 = pd.concat([producto1, producto2, producto3, producto4], axis=1)
consolidado2.columns = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
#print(consolidado2)
#consolidado2['Tienda 1'].idxmax()
for x in consolidado2.columns:
    valor_maximo=consolidado2[x].max()
    filas_maximas = consolidado2[consolidado2[x] == valor_maximo]
    indices_maximos = filas_maximas.index.tolist()
    print(f' el producto mas vendido de la {x} es o son:  {indices_maximos}')
    valor_minimo=consolidado2[x].min()
    filas_minimas = consolidado2[consolidado2[x] == valor_minimo]
    indices_minimos = filas_minimas.index.tolist()
    print(f' el producto menos vendido de la {x} es o son:  {indices_minimos}\n')


consolidado2
```

```python
# Suponiendo que tienes los DataFrames tienda1, tienda2, tienda3, tienda4
# con la columna 'Producto' que contiene el nombre de cada producto

# Combinar los DataFrames de las tiendas
all_stores = pd.concat([tienda1, tienda2, tienda3, tienda4], keys=['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'])
# Reiniciar el índice para evitar el error 'level_0'
all_stores = all_stores.reset_index()
# Agrupar por tienda y producto y contar las ventas
product_counts = all_stores.groupby(['level_0', 'Producto'])['Producto'].count().reset_index(name='Ventas')

# Crear la figura y los subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Iterar sobre las tiendas y los subplots
for i, store in enumerate(['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']):
    row = i // 2  # Fila del subplot
    col = i % 2   # Columna del subplot
    ax = axs[row, col]

    # Filtrar los datos de la tienda actual
    store_data = product_counts[product_counts['level_0'] == store]

    # Crear el gráfico de barras
    ax.bar(store_data['Producto'], store_data['Ventas'])

    # Configurar el título y las etiquetas
    ax.set_title(store)
    ax.set_xlabel('Producto')
    ax.set_ylabel('Ventas')

    # Rotar las etiquetas del eje x para mejor legibilidad
    plt.setp(ax.get_xticklabels(), rotation=85, ha='right')

# Ajustar el espaciado entre los subplots
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python

!pip install cartopy
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import pandas as pd

# Crear la figura y los subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 8), subplot_kw={'projection': ccrs.PlateCarree()})

# Iterar sobre las tiendas y los subplots
for i, tienda in enumerate([tienda1, tienda2, tienda3, tienda4]):
    row = i // 2  # Fila del subplot
    col = i % 2   # Columna del subplot
    ax = axs[row, col]

    # Agrupar por latitud, longitud y sumar las ventas para la tienda
    ventas_por_region = tienda.groupby(['lat', 'lon'])['Precio'].sum().reset_index()

    # Agregar características geográficas al mapa
    ax.add_feature(cfeature.COASTLINE)
    ax.add_feature(cfeature.BORDERS, linestyle=':')
    ax.add_feature(cfeature.LAND)
    ax.add_feature(cfeature.OCEAN)

    # Crear el diagrama de dispersión
    scatter = ax.scatter(ventas_por_region['lon'], ventas_por_region['lat'],
                         transform=ccrs.PlateCarree(),
                         s=ventas_por_region['Precio'] * 0.00001,  # Ajustar el factor de escala
                         alpha=0.5,
                         c='blue')

    # Agregar una barra de color para indicar el tamaño de los puntos (solo en el primer subplot)
    if i == 0:
        plt.colorbar(scatter, ax=ax, label='Ventas')

    # Configurar título y etiquetas
    ax.set_title(f'Tienda {i + 1}')
    ax.set_xlabel('Longitud')
    ax.set_ylabel('Latitud')

# Ajustar el espaciado entre los subplots
plt.tight_layout()

# Mostrar el gráfico
plt.show()

# Crear la figura y los subplots
fig, axs = plt.subplots(2, 2, figsize=(16, 14))

# Iterar sobre las tiendas y los subplots
for i, tienda in enumerate([tienda1, tienda2, tienda3, tienda4]):
    row = i // 2  # Fila del subplot
    col = i % 2   # Columna del subplot
    ax = axs[row, col]

    # Obtener las ventas por lugar de compra para la tienda actual
    ventas_por_lugar = tienda['Lugar de Compra'].value_counts()

    # Crear el gráfico de pastel
    ax.pie(ventas_por_lugar, labels=ventas_por_lugar.index, autopct='%1.1f%%', startangle=180,rotatelabels=True, textprops={'fontsize': 10})

    # Configurar el título con el número de tienda
    ax.set_title(f'Tienda {i + 1}')

# Ajustar el espaciado entre los subplots
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```
